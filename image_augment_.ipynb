```python
# Step 1: Load your local image
img = load_img('exp5.jpg', target_size=(224, 224))  # Resize to a standard size
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
plt.show()
```

```python

# Step 2: Convert image to array and reshape
img_array = img_to_array(img)
img_array = img_array.reshape((1,) + img_array.shape)  # Shape: (1, 224, 224, 3)
```

```python
# Step 3: Define the ImageDataGenerator with augmentation settings
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
# Step 4: Generate augmented images
aug_iter = datagen.flow(img_array, batch_size=1)

# Step 5: Plot 6 augmented images
plt.figure(figsize=(12, 6))
for i in range(6):
    batch = next(aug_iter)
    image = batch[0].astype('uint8')  # For correct display
    plt.subplot(2, 3, i + 1)
    plt.imshow(image)
    plt.axis('off')

plt.suptitle("Augmented Images from exp5.jpg")
plt.tight_layout()
plt.show()
```
